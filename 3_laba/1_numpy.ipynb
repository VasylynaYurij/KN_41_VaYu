### Метод linspace - генерує задану кількістю чисел у певному діапазоні, 50 чисел за замовчуванням

```python
a = np.linspace(start, size_of_vec)
print(f"Згенерували {len(a)} чисел в діапазоні {a[0]} до {a[-1]}")

# Вивести кожне друге число згенерованого діапазону за допомогою циклу
print("Кожне друге число згенерованого діапазону:")
for i in range(1, len(a), 2):
    print(a[i])

print("Можна зменшити кількість чисел до 5:")
a = np.linspace(start, size_of_vec, 5)
print(a)
# Збільшіть кількість чисел до 100. Виведіть перших та останніх 10.
a = np.linspace(start, size_of_vec, 100)
print(f"Перші 10 чисел: {a[:10]}")
print(f"Останні 10 чисел: {a[-10:]}")

print("""Як бачимо перше та останнє значення завжди входять у згенерований діапазон. 
Відповідно при цьому вираховується крок.
Це можна змінити, виключивши останнє значення з визначення діапазону.
""")

print(np.linspace(start, size_of_vec, 5, endpoint=False))
# Спробуйте самі порахувати крок.
print("Крок у linspace без включення кінцевого значення:", np.linspace(start, size_of_vec, 5, endpoint=False)[1] - np.linspace(start, size_of_vec, 5, endpoint=False)[0])

_, sp_1 = np.linspace(start, size_of_vec, 5, endpoint=True, retstep=True)
_, sp_2 = np.linspace(start, size_of_vec, 5, endpoint=False, retstep=True)
print(f"Для порівняння, ми можемо визначити крок автоматично, у нашому випадку це {sp_1} та {sp_2}")
# Чому маємо такий крок?
print(f"Крок в першому випадку (з endpoint=True): {sp_1}, а в другому випадку (з endpoint=False): {sp_2}. Крок змінюється через те, чи включено кінцеве значення в діапазон.")
```

### Типи масивів/векторів

```python
import numpy as np

examples = {
    'Скаляр': np.array(1),
    'Вектор Рядок': np.array([1, 2, 3]),
    'Вектор Стовпець': np.array([[1], [2], [3]]),
    'Матриця': np.array([[1, 2], [3, 4], [5, 6]]),
    'Багатовимірна Матриця': np.array([[[111, 112], [121, 122]],
                                       [[211, 212], [221, 222]],
                                       [[311, 312], [321, 322]]])
}

for key in examples:
    print(f"Розглядаємо приклад з {key}. \nЗначення x буде: \n", examples[key])
    print(f"Тип {key} x: ", type(examples[key]))
    print(f"Розмірність/вимір {key} x:", np.ndim(examples[key]))
    print(f"Форма {key} x: ", examples[key].shape)
    print(f"Кількість елементів х: {np.size(examples[key])}")
    print("-=-" * 10)

# Homework
homework("""
Як Ви думаєте чому маємо такий результат для функції np.ndim (Розмірність/вимір/dimension)?
Чому значення для векторів та матриць однакове та не співпадає з значенням shape?
Задайте власні вектори з різними числами.
""")
# Відповідь на запитання
print("Відповідь на запитання: np.ndim повертає кількість осей масиву. Для скаляра це 0, для вектора 1, для матриці 2 і т.д.\n Значення np.ndim для векторів та матриць однакове, тому що функція просто підраховує кількість вимірів,\n а не кількість елементів у масиві. Вектор (рядок) і вектор (стовпець) мають по одному виміру, але їх форма різна.")

# Приклад власних векторів:
vektor1 = np.array([10, 20, 30])  # Вектор 1
vektor2 = np.array([[5], [10]])  # Вектор 2 (стовпець)
print("Тип вектора1:", np.ndim(vektor1), "Форма:", vektor1.shape)
print("Тип вектора2:", np.ndim(vektor2), "Форма:", vektor2.shape)
```
